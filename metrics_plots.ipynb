```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all three CSVs
hkad_cache = pd.read_csv("store_metrics_hkademlia_with_caching.csv")
hkad = pd.read_csv("store_metrics_hkademlia.csv")
kad_cache = pd.read_csv("store_metrics_kademlia_with_caching.csv")
kad = pd.read_csv("store_metrics_kademlia.csv")


# Plotting
ticks = hkad["Tick"]

fig, axs = plt.subplots(1, 3, figsize=(18, 5), sharex=True)

# Plot 1: Store Hops
axs[0].plot(ticks, hkad_cache["AvgStoreHops"], marker="s", label="Hkad+cache")
axs[0].plot(ticks, hkad["AvgStoreHops"], marker="D", label="Hkad")
axs[0].plot(ticks, kad_cache["AvgStoreHops"], marker="^", label="Kad+cache")
axs[0].plot(ticks, kad["AvgStoreHops"], marker="o", label="Kad")
axs[0].set_ylabel("Store hops")
axs[0].set_xlabel("Experiment Progress (150k requests/tick)")
axs[0].legend()
axs[0].grid(True)

# Plot 2: Store Latency
axs[1].plot(ticks, hkad_cache["AvgStoreLatency(ms)"], marker="s", label="Hkad+cache")
axs[1].plot(ticks, hkad["AvgStoreLatency(ms)"], marker="D", label="Hkad")
axs[1].plot(ticks, kad_cache["AvgStoreLatency(ms)"], marker="^", label="Kad+cache")
axs[1].plot(ticks, kad["AvgStoreLatency(ms)"], marker="o", label="Kad")
axs[1].set_ylabel("Store latency (ms)")
axs[1].set_xlabel("Experiment Progress (150k requests/tick)")
axs[1].legend()
axs[1].grid(True)

# Plot 3: Store Receivers
axs[2].plot(ticks, hkad_cache["StoreReceivers"], marker="s", label="Hkad+cache")
axs[2].plot(ticks, hkad["StoreReceivers"], marker="D", label="Hkad")
axs[2].plot(ticks, kad_cache["StoreReceivers"], marker="^", label="Kad+cache")
axs[2].plot(ticks, kad["StoreReceivers"], marker="o", label="Kad")
axs[2].set_ylabel("Store receivers")
axs[2].set_xlabel("Experiment Progress (150k requests/tick)")
axs[2].legend()
axs[2].set_ylim(19, 20.25)
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all three CSVs
hkad_cache = pd.read_csv("InterToIntra_hk_cache.csv")
hkad = pd.read_csv("InterToIntra_hk.csv")
kad_cache = pd.read_csv("InterToIntra_k_cache.csv")
kad = pd.read_csv("InterToIntra_k.csv")


# Plotting three subplots: KBucket Size, Store Inter/Intra Ratio, Lookup Inter/Intra Ratio
fig, axs = plt.subplots(1, 3, figsize=(18, 5), sharex=True)
ticks = hkad["Tick"]

# Plot 1: KBucket Size
axs[0].plot(ticks, hkad_cache["KBucket Size Δ"], marker="s", label="Hkad+cache")
axs[0].plot(ticks, hkad["KBucket Size Δ"], marker="D", label="Hkad")
axs[0].plot(ticks, kad_cache["KBucket Size Δ"], marker="^", label="Kad+cache")
axs[0].plot(ticks, kad["KBucket Size Δ"], marker="o", label="Kad")
axs[0].set_ylabel("KBucket Size (peers)")
axs[0].set_xlabel("Experiment Progress (150k requests/tick)")
axs[0].set_title("KBucket Size over Time")
axs[0].legend()
axs[0].grid(True)

# Plot 2: Store Inter/Intra Cluster Ratio
axs[1].plot(ticks, hkad_cache["InterToIntraCluster Ratio - Store"], marker="s", label="Hkad+cache")
axs[1].plot(ticks, hkad["InterToIntraCluster Ratio - Store"], marker="D", label="Hkad")
axs[1].plot(ticks, kad_cache["InterToIntraCluster Ratio - Store"], marker="^", label="Kad+cache")
axs[1].plot(ticks, kad["InterToIntraCluster Ratio - Store"], marker="o", label="Kad")
axs[1].set_ylabel("InterToIntraCluster Ratio - Store")
axs[1].set_xlabel("Experiment Progress (150k requests/tick)")
axs[1].set_title("Store Inter/Intra Ratio")
axs[1].legend()
axs[1].grid(True)

# Plot 3: Lookup Inter/Intra Cluster Ratio
axs[2].plot(ticks, hkad_cache["InterToIntraCluster Ratio - Lookup"], marker="s", label="Hkad+cache")
axs[2].plot(ticks, hkad["InterToIntraCluster Ratio - Lookup"], marker="D", label="Hkad")
axs[2].plot(ticks, kad_cache["InterToIntraCluster Ratio - Lookup"], marker="^", label="Kad+cache")
axs[2].plot(ticks, kad["InterToIntraCluster Ratio - Lookup"], marker="o", label="Kad")
axs[2].set_ylabel("InterToIntraCluster Ratio - Lookup")
axs[2].set_xlabel("Experiment Progress (150k requests/tick)")
axs[2].set_title("Lookup Inter/Intra Ratio")
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
